# Projectile Motion Problem **solver**

```python
from math import sin, cos, tan, radians
from sympy import*
from numpy import*
x,y,z=symbols('x,y,z')
from matplotlib.pyplot import*
import matplotlib.animation as animation
g=9.8
def Max_Height(u,theta):
    Hmax=(u*sin(radians(theta)))**2/(2*g)
    return Hmax
def Time_of_flight(u,theta):
    T=(2*u*sin(radians(theta)))/g
    return f"Time of flight of the Projectile is {round(T,4)} seconds"
def Horizontal_range(u,theta):
    R=(u**2*sin(radians(2*theta))/g)
    return R

def Equation_of_path(u,theta):
    func=(x*tan(radians(theta)))-((g*x**2)/(2*u**2*cos(radians(theta))**2))
    X_co= Horizontal_range(u,theta)
    X=linspace(0,X_co,200)
    Y= (X*tan(radians(theta)))-((g*X**2)/(2*u**2*cos(radians(theta))**2))
    plot(X,Y)
    grid()
    show()
    def anim():

        x_p = linspace(0,X_co,1000)
        y_p = (x_p*tan(radians(theta)))-((g*x_p**2)/(2*u**2*cos(radians(theta))**2))

        fig, ax = subplots()
        ax.set_xlim([0,X_co+5])
        ax.set_ylim([0,Max_Height(u,theta)+5])
        line, = ax.plot(   [],[], lw=2,  color='black')
        sctr  = ax.scatter([],[], s=100, color='orange')

        def update(i):
            line.set_ydata(y_p[:i+1])  # update
            line.set_xdata(x_p[:i+1])
            sctr.set_offsets((x_p[i],y_p[i]))
            return line,sctr
        ani = animation.FuncAnimation(fig, update, 1000, interval=5, blit=True)
        ani.save('plotanimation.gif')
        grid()
        show()

    anim()

    return f"Equation of path of projectile motion is {func}\nMaximum Height of the Projectile {Max_Height(u,theta)}\nand its range is\n{X_co} meters"
def position(u,theta):
    time=abs(float(input(f"enter a time lower than {Time_of_flight(u,theta)}")))
    x_cordi=(u*cos(radians(theta)))*time
    y_cordi=((u*sin(radians(theta)))*time)-(0.5*g*time**2)
    X_co= Horizontal_range(u,theta)
    X=linspace(0,X_co,100)
    Y= (X*tan(radians(theta)))-((g*X**2)/(2*u**2*cos(radians(theta))**2))
    plot(X,Y, color='teal')
    scatter([x_cordi],[y_cordi], color='purple', zorder=5)
    grid()
    show()
    return f"X coordinate = {(x_cordi)}, & Y coordinate= {y_cordi}"
def wall(u,theta,Wal_height,wal_distance):
    y_cordi=(wal_distance*tan(radians(theta)))-((g*wal_distance**2)/(2*u**2*cos(radians(theta))**2))
    if y_cordi<=Wal_height:
        return f"will hit the wall"
    else:
        return f"won' hit the wall"
def Area_under(u, theta):
  func=(x*tan((theta)))-((g*x**2)/(2*u**2*cos((theta))**2))

  return round(integrate(func,(x,0,Horizontal_range(u,theta))),4)
#print(wall(20,70,10,20))
# print(Time_of_flight(70,35))
# print(Equation_of_path(35,70))
```
